```python

import numpy as np
import pandas as pd
import os
#os.chdir('C:\\Users\\caneu\\Downloads')
sample_data = pd.read_excel('past_four_womens_data.xlsx')
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
sample_data.head()
```

```python
#converting to year
sample_data['season_id'] = sample_data['season_id'] + 1364
```

```python
sample_data['player&season'] = sample_data['player.name'] +' '+ sample_data['season_id'].astype('string')
```

# Using only relevant columns

```python
sample_data = sample_data[[
       'player.id','player.normalized','player&season', 'boxscore.pts',
        'boxscore.rebso', 'boxscore.rebsd', 'boxscore.ast',
       'boxscore.rebs','boxscore.fta','boxscore.fga',
       'boxscore.to', 'calculated.pps_noft',
       'calculated.true_shooting', 'calculated.usage','calculated.atr_attempts',
       'calculated.atr_rate','calculated.three_rate','calculated.three_attempts',
       'calculated.mid_rate', 'calculated.mid_attempts','calculated.assisted_percentage','calculated.possessions']]
```

# Converting to stats per 100

```python
for num in range(0,sample_data.shape[0]):
   sample_data.loc[num,['boxscore.pts','boxscore.rebso', 'boxscore.rebsd', 'boxscore.ast','boxscore.rebs','boxscore.fta','boxscore.fga','boxscore.to'] ] = sample_data[['boxscore.pts','boxscore.rebso', 'boxscore.rebsd', 'boxscore.ast','boxscore.rebs','boxscore.fta','boxscore.fga','boxscore.to']].loc[num].div(sample_data['calculated.possessions'].loc[num]).mul(100)
```

# Creating points per shot stat

```python
for num in range(0, sample_data.shape[0]):
    #sample_data.shape[0]
    #.div(sample_data.loc[num, ['calculated.atr_attempts']])
    sample_data.loc[num, 'calculated.atr_attempts'] = sample_data.loc[num, 'calculated.atr_attempts'] * (sample_data.loc[num, 'calculated.atr_rate']) * 2 / sample_data.loc[num, 'calculated.atr_attempts'] 
```

```python
for num in range(0, sample_data.shape[0]):
    #sample_data.shape[0]
    #.div(sample_data.loc[num, ['calculated.atr_attempts']])
    sample_data.loc[num, 'calculated.mid_attempts'] = sample_data.loc[num, 'calculated.mid_attempts'] * (sample_data.loc[num, 'calculated.mid_rate']) * 2 / sample_data.loc[num, 'calculated.mid_attempts'] 
```

```python
for num in range(0, sample_data.shape[0]):
    #sample_data.shape[0]
    #.div(sample_data.loc[num, ['calculated.atr_attempts']])
    sample_data.loc[num, 'calculated.three_attempts'] = sample_data.loc[num, 'calculated.three_attempts'] * (sample_data.loc[num, 'calculated.three_rate']) * 3 / sample_data.loc[num, 'calculated.three_attempts'] 
```

# Creating shot type per 100 stat

```python
for num in range(0,sample_data.shape[0]):
    sample_data.loc[num, ['calculated.atr_rate', 'calculated.three_rate', 'calculated.mid_rate']] = sample_data.loc[num, ['calculated.atr_rate', 'calculated.three_rate', 'calculated.mid_rate']].mul(sample_data.loc[num, 'boxscore.fga'])
```

# Renaming all columns

```python
sample_data.columns = ['player.id', 'player.name','player&season', 'ppp_100', 'oreb_100',
       'dreb_100', 'ast_100', 'tot_reb_100', 'fta_100',
       'fga_100', 'tov_100', 'pps_noft',
       'true_shooting', 'usage','rim_pps', 'rim_100',
       'three_100', 'three_pps','mid_100', 'mid_pps',
       'assisted_percentage', 'possessions']
```
